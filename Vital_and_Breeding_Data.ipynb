```python
from matplotlib import pyplot as plt
import numpy as np

index=[1,2,3,4,5,6,7,8]
stress=[8.6, 3.6, 1.3, 1.8, 2.7, 3.2, 8.9, 1.3]
exercise=[2,3,2,3,1,2,2,2]
food_count=[4,4,2,2,2,4,4,2]
```

```python
plt.scatter(exercise, stress)
plt.xlabel("exercise")
plt.ylabel("stress")
```

```python
plt.scatter(food_count, stress)
plt.xlabel("food_count")
plt.ylabel("stress")
```

```python
plt.scatter(exercise, stress, color="skyblue", label="exercise")
plt.scatter(food_count, stress, color="pink", label="food_count")
plt.ylabel("stress")
plt.legend(loc="best")
```

```python
plt.scatter(index, exercise, color="skyblue", label="exercise")
plt.scatter(index, food_count, color="lightgreen", label="food_count")
plt.scatter(index, stress, color="pink", label="stress")
plt.xlabel("index")
plt.legend(loc="best")
stress_mean=float(vital_df['stress'].mean())
exercise_mean=float(breed_df['exercise'].mean())
foodcount_mean=float(breed_df['food-count'].mean())
plt.axhline(stress_mean,label='Mean', color='r') ## 평균값을 y좌표로 하는 수평선 생성
plt.axhline(exercise_mean,label='Mean', color='b') ## 평균값을 y좌표로 하는 수평선 생성
plt.axhline(foodcount_mean,label='Mean', color='g') ## 평균값을 y좌표로 하는 수평선 생성
```
